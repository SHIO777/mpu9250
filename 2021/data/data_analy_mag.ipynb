```python
# import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
from matplotlib import animation
from IPython.display import HTML
```

```python
filename1 = "udp_log_20210915-151429.csv"
df = pd.read_csv(filename1, delimiter=",", encoding="cp932")
df.head()
```

```python
df['yaw_calc'] = 0.0
df['yaw_deg'] = 0.0
df['yaw_x'] = 0.0
df['yaw_y'] = 0.0
r = 1.0

for i in range(len(df)):
    df['yaw_calc'][i] = "{:.3f}".format(math.atan2(df['mag[1]'][i], df['mag[0]'][i]))
    df["yaw_deg"][i] = math.degrees(df["yaw"][i])
    df['yaw_x'][i] = r * math.cos(df['yaw'][i])
    df['yaw_y'][i] = r * math.sin(df['yaw'][i])
# df["diff"] = df['yaw'] - df['yaw_calc']
df
```

```python
df2 = df.reindex(columns=["pitch", "roll", "mag[2]", "mag[0]", "mag[1]", "yaw", "yaw_deg", "yaw_calc", "diff"])
df2
```

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(111, xlim=(-1.2, 1.2), ylim=(-1.2, 1.2))
# ax = fig.add_subplot(211, xlim=(0, 500), ylim=(-2.0, 2.0))

nframes = len(df)

def init():
    return(ax)

def animate(i):
    # ax.plot(df['yaw_x'][i], df['yaw_y'][i], marker=".")
    # ax.plot(df['yaw_x'][i], df['yaw_y'][i])
    ax.scatter(df['yaw_x'][i], df['yaw_y'][i])
    return(ax)


anim= animation.FuncAnimation(fig, animate, init_func=init, frames = nframes, interval = 50)
plt.close()
HTML(anim.to_jshtml())
# ax.legend()
# ax.set_title("Gyroscope Calibration Offset Correction", fontsize=22)
# ax.set_xlabel("", fontsize=18)
# ax.set_ylabel("$\omega_{x, y, z}$ [$^{\circ}/sec$]", fontsize=18)
# ax2.set_xlabel("Sample", fontsize=18)
# ax2.set_ylabel("$\omega_{x, y, z}$ [$^{\circ}/sec$]", fontsize=18)

# plt.savefig('gyro.png', dpi=500)
# plt.close('all')
```

```python


filename1 = "2_uncalibrated.csv"
filename2 = "2_calibrated.csv"


df = pd.read_csv(filename1, delimiter=",", encoding="cp932")
# df.head()
df2 = pd.read_csv(filename2, delimiter=",", encoding="cp932")
# df2.head()

fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(211, xlim=(0, 500), ylim=(-2.0, 2.0))
ax2 = fig.add_subplot(212, xlim=(0, 500), ylim=(-2.0, 2.0))

ax.plot(df['x [dps]'], label="$\omega_x, Uncalibrated$")
ax.plot(df['y [dps]'], label="$\omega_y, Uncalibrated$")
ax.plot(df['z [dps]'], label="$\omega_z, Uncalibrated$")

ax2.plot(df2['x [dps]'], label="$\omega_x, Calibrated$")
ax2.plot(df2['y [dps]'], label="$\omega_y, Calibrated$")
ax2.plot(df2['z [dps]'], label="$\omega_z, Calibrated$")

ax.legend()
ax2.legend()
ax.set_title("Gyroscope Calibration Offset Correction", fontsize=22)
ax.set_xlabel("", fontsize=18)
ax.set_ylabel("$\omega_{x, y, z}$ [$^{\circ}/sec$]", fontsize=18)
ax2.set_xlabel("Sample", fontsize=18)
ax2.set_ylabel("$\omega_{x, y, z}$ [$^{\circ}/sec$]", fontsize=18)

plt.savefig('gyro.png', dpi=500)
plt.close('all')
```
